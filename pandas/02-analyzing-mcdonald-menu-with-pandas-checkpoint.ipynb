### เมนูไหนมีแคลอรี่เยอะที่สุด?

```python
import pandas as pd
```

```python
df = pd.read_csv("mcdonald-menu.csv")
```

```python
df.head()
```

```python
df.sort_values(by="Calories from Fat", ascending=False)
```

```python
idx = df.Calories.idxmax()
```

```python
df.loc[idx, ['Item', 'Category', 'Calories']]
```

### เมนูไหนมีแคลอรี่น้อยที่สุด?

```python
df.sort_values(by="Calories from Fat", ascending=True)
```

```python
idx = df.Calories.idxmin()
```

```python
df.loc[idx, ['Item', 'Category', 'Calories']]
```

### โดยเฉลี่ยแล้วเมนูแต่ละประเภทมีแคลอรี่เท่าไหร่?

```python
df.groupby('Category')['Calories'].mean()
```

### ถ้าสั่งแซนวิชไก่ย่าง (grilled chicken) จะได้สารอาหารดีกว่าสั่งแซนวิชไก่ทอด (crispy chicken) หรือเปล่า?

```python
def check_grilled_chicken(value) :
    if "Grilled Chicken" in value :
        return True
    else : return False
```

```python
df['Grilled'] = df.Item.map(check_grilled_chicken)
```

```python
df[df.Grilled == True].Calories.mean()
```

```python
def check_crispy(value) :
    if "Crispy" in value :
        return True
    else : return False
```

```python
df["Crispy"] = df.Item.map(check_crispy)
```

```python
df[df.Crispy == True].Calories.mean()
```

### ถ้าสั่งเมนูที่มีแต่ไข่ขาว (egg whites) จะได้สารอาหารดีกว่าเมนูที่ใช้ทั้งไข่ขาวกับไข่แดงหรือเปล่า?

```python
def check_egg(value) :
    if "egg whites" in value :
        return True
    else : return False
```

```python
df["egg whites"] = df.Item.map(check_egg)
```

```python
df[df.egg == True].Calories.mean()
```
